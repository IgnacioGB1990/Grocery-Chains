## Step 1: Clean Mercadona csv


```python
df = pd.read_csv("./data/mercadona.csv")
original_df = df.count()
print(original_df)
df.head()
```

### Price format

```python
# Replace , with . , splits into array, gets first number 
df["Price"] = df["Price"].apply(lambda x: x.replace(",",".").split(" ")[0])
df["Price"].head()
```

```python
# Shows all values that are strings 
df[df["Price"].apply(lambda x: "." not in x)].head()
```

```python
#Using drop() to delete rows based on column value
df.drop(df[df["Price"].apply(lambda x: "." not in x)].index,inplace =True)
df.drop(df[df["Price"].apply(lambda x: len(x)>6)].index,inplace =True)
```

```python
print("Original DF",original_df[0])
print("After",df.count()[0])
df.head()
```

```python
# Append size to Name ( de momento )
df["Name_Size"]= df["Name"] + " " + df["Size"]

df
```

### Save cleaned format as csv and excel (Prices as float, more to do)

```python
format_mercadona = df[["Shop","Group","Name_Size","Price","Date"]]
format_mercadona
```

```python
format_mercadona = format_mercadona.rename(columns={"Group":"Category","Name_Size":"Name"})
format_mercadona
```
