## Step 1: Clean Mercadona csv


```python
df = pd.read_csv("./data/mercadona.csv")
print(df.count())
df.head()
```

### Price format

```python
# Replace , with . , splits into array, gets first number 
df["Price"] = df["Price"].apply(lambda x: x.replace(",",".").split(" ")[0])
df["Price"].head()
```

```python
# Shows all values that are strings 
df[df["Price"].apply(lambda x: "." not in x)].head()
```

```python
#Using drop() to delete rows based on column value
df.drop(df[df["Price"].apply(lambda x: "." not in x)].index,inplace =True)
df.drop(df[df["Price"].apply(lambda x: len(x)>6)].index,inplace =True)
```
