## Step 1: Clean Alcampo csv

```python
df = pd.read_csv("./data/alcampo.csv")
original_df = df.count()
print(original_df)

# Change to String first
df["Price_unit"] = df["Price"].astype(str)
df.tail()
```

```python
# All , to . and separate each characters
df["Price_unit"] = df["Price_unit"].apply(lambda x: x.replace(",",".").split("€"))
df
```

```python
df.drop(df[df["Price_unit"].apply(lambda x: "(" in x[0])].index,inplace =True)
df
```

```python
df["Price"] = df["Price_unit"].apply(lambda x: x[0]).astype(float)
df
```

```python
# Price check
df[df["Price"]>10]
```
