```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("./mercadona.csv",parse_dates=["Date"])
df1.head()
```

```python

df2 = pd.read_csv("./alcampo.csv",parse_dates=["Date"])
df2.head()
```

```python
df3 = pd.concat([df1, df2])
df3
```
